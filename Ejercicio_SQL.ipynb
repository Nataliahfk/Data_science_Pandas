```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
import pandas as pd
```

```python
archivo = 'clientes_banco.csv'
```

```python
datos = pd.read_csv(archivo)
datos
```

```python
# aqui estamos creando la tabla clientes y llamamos al motor antes guardado en resultado_engine
datos.to_sql('clientes', engine , index=False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
pd.read_sql_table('clientes', con = engine.connect())
```

Realizar tres actualizaciones en la base de datos:

- Actualizar el registro del cliente ID 6840104 cuyo rendimiento anual cambió a 300000.

```python
from sqlalchemy.exc import SQLAlchemyError # este comando muestra que tipo de error, me entrega información del error
query = 'UPDATE clientes SET Rendimiento_anual = 300000 WHERE ID_Cliente = 6840104 '
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('#Registros actualizados: ', r_set.rowcount )
```

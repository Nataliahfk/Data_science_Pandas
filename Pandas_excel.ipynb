###**Aula 2 - Leyendo archivos Excel**

###**Importando archivos Excel**

```python
import pandas as pd
```

```python
archivo = 'emisiones_CO2.xlsx'
```

```python
# leo y convierto en DF el excel
datos_co2 = pd.read_excel(archivo)
datos_co2
```

Microsoft Excel es una de las herramientas de hojas de cálculo más utilizadas en el mundo, siendo ampliamente utilizada para almacenar y analizar datos en formato de tabla. En el video anterior, mostramos cómo la función read_excel se puede usar para leer un archivo de Excel en formato xlsx. Sin embargo, esta función es capaz de leer archivos en otros formatos, como: xls, xlsm, xlsb, odf, ods y odt.

```python
datos_co2.head()
```

```python
# con este comando pido que me muestre el nombre de las hojas
pd.ExcelFile(archivo).sheet_names
```

###**Parámetros de la función read_excel**
Documentación: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html#pandas.read_excel

```python
# Para ver las otras ojas utilizaremos la f(x) sheetname
emision_percapita = pd.read_excel(archivo, sheet_name='emisiones_percapita')
emision_percapita.tail() # muestra los 5 últimos registros
```

```python
hoja_fuentes = pd.read_excel(archivo, sheet_name='fuentes')
hoja_fuentes.sample(8) # muestra 8 registros de forma aleatoria
```
